```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

```python
df = pd.read_csv("winequality-white.csv", sep = ";")
df.head()
```

```python
help(sns.histplot)
```

```python
sns.histplot( x = "volatile acidity", data=df)
```

```python
sns.set(style="whitegrid")
ax = sns.scatterplot(y = "pH", x = df.index  , data=df, size = 2)
ax.set(xlabel = 'Number of sampling times', ylabel = 'pH')
plt.title('My first graph')
plt.show()
```

```python
sns.set(style="whitegrid")
ax = sns.scatterplot(y = "citric acid", x = df.index  , data=df, size = 2)
ax.set(xlabel = 'Number of sampling times', ylabel = 'Citric acid')
plt.title('The second graph')
plt.show()
```

```python
sns.catplot(x = "pH", data=df, aspect=10, kind="count", palette = 'deep' )
# thng k s ln xut hin
```

```python
sns.regplot(y='pH', x=df.index, data=df, marker="+")
```
