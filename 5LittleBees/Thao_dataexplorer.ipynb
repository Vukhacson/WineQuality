```python
#Correlation
corr_red = red_wine.corr()
sns.heatmap(corr_red)
```

```python
#Covariance
cov_red = red_wine.cov()
sns.heatmap(cov_red)
```

```python
#Info and type
red_wine['fixed acidity'].info()
red_wine['volatile acidity'].info()
red_wine['citric acid'].info()
red_wine['residual sugar'].info()
red_wine['chlorides'].info()
red_wine['free sulfur dioxide'].info()
red_wine['total sulfur dioxide'].info()
red_wine['density'].info()
red_wine['pH'].info()
red_wine['sulphates'].info()
red_wine['alcohol'].info()
red_wine['quality'].info()
```

```python
#Visualize(grid map 2x6)

fix_acid = red_wine['fixed acidity']
vol_acid = red_wine['volatile acidity']
cit_acid = red_wine['citric acid']
res_sur = red_wine['residual sugar']
clo = red_wine['chlorides']
fsd = red_wine['free sulfur dioxide']
tsd = red_wine['total sulfur dioxide']
dens = red_wine['density']
pH = red_wine['pH']
sul = red_wine['sulphates']
alc = red_wine['alcohol']
qual = red_wine['quality']

fig, axs = plt.subplots(6, 2, figsize=(20, 12), constrained_layout = True)
axs[0, 0].hist(fix_acid, bins=200)
axs[0, 0].set_title("Fixed acidity")
axs[0, 1].hist(vol_acid, bins=200)
axs[0, 1].set_title("Volatile acidity")
axs[1, 0].hist(cit_acid, bins=200)
axs[1, 0].set_title("Citric acid")
axs[1, 1].hist(res_sur, bins=200)
axs[1, 1].set_title("Residual sugar")
axs[2, 0].hist(clo, bins=200)
axs[2, 0].set_title("Chlorides")
axs[2, 1].hist(fsd, bins=200)
axs[2, 1].set_title('Free sulfur dioxide')
axs[3, 0].hist(tsd, bins=200)
axs[3, 0].set_title('Total sulfur dioxide')
axs[3, 1].hist(dens, bins=200)
axs[3, 1].set_title('Density')                    
axs[4, 0].hist(pH, bins=200)
axs[4, 0].set_title("pH")                   
axs[4, 1].hist(sul, bins=200)
axs[4, 1].set_title('Sulphates')                    
axs[5, 0].hist(alc, bins=200)
axs[5, 0].set_title('Alcohol')                    
axs[5, 1].hist(qual, bins=200)
axs[5, 1].set_title("Quality")
```
