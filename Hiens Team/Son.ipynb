```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
path1 = "winequality-white.csv"
path2 = "winequality-red.csv"

white_df = pd.read_csv(path1, delimiter=";")
red_df = pd.read_csv(path2, delimiter=";")
combine = pd.concat([white_df, red_df])
```

```python
white_df.corr()['quality']
```

```python
red_df.corr()['quality']
```

```python
corr = combine.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True, cmap="YlGnBu", linewidths=0.1, annot_kws={"fontsize":12})
plt.title("Correlation of Features")
```

```python
white_df.cov()['quality']
```

```python
red_df.cov()['quality']
```

```python
cov = combine.cov()
plt.figure(figsize=(15, 10))
sns.heatmap(cov, annot=True, cmap="RdPu", linewidths=0.1, annot_kws={"fontsize":12})
plt.title("Covariance of Feature")
```

```python
fig,ax=plt.subplots(6,2,figsize=(15,30))
sns.countplot(x=combine.quality,ax=ax[0][0]).set_title('Target Count',size=15)
sns.boxplot(x=combine.quality,y=combine['fixed acidity'],ax=ax[0][1])
sns.boxplot(x=combine.quality,y=combine['citric acid'],ax=ax[1][0])
sns.boxplot(x=combine.quality,y=combine['residual sugar'],ax=ax[1][1])
sns.boxplot(x=combine.quality,y=combine['chlorides'],ax=ax[2][0])
sns.boxplot(x=combine.quality,y=combine['free sulfur dioxide'],ax=ax[2][1])
sns.boxplot(x=combine.quality,y=combine['total sulfur dioxide'],ax=ax[3][0])
sns.boxplot(x=combine.quality,y=combine['density'],ax=ax[3][1])
sns.boxplot(x=combine.quality,y=combine['pH'],ax=ax[4][0])
sns.boxplot(x=combine.quality,y=combine['sulphates'],ax=ax[4][1])
sns.boxplot(x=combine.quality,y=combine['alcohol'],ax=ax[5][0])
sns.boxplot(x=combine.quality,y=combine['volatile acidity'],ax=ax[5][1])
```
