# 1.2 Check missing data point, plot a heat map of showing if null is exist

```python
df_red_wine_nulls = df_red_wine.isnull()
df_white_wine_nulls = df_white_wine.isnull()
print("Check total missing data points of red wine data: ", df_red_wine_nulls.sum().sum())
print("Check total missing data points of white wine data: ", df_white_wine_nulls.sum().sum())
```

```python
# Plot heatmap of null values
plt.imshow(df_red_wine_nulls, cmap="Reds")
plt.show()
```

# 5. Correlation and Covariance

```python
df_red_cov_matrix = df_red_wine.cov()
df_white_cov_matrix = df_white_wine.cov()
df_red_corr_matrix = df_red_wine.corr()
df_white_corr_matrix = df_white_wine.corr()
```

```python
df_white_cov_matrix.head()
```

```python
import seaborn as sns
plt.figure(figsize = (14,8))
plt.title("Correlation matrix of wine features")
sns.heatmap(df_red_corr_matrix, annot = True)
```

# 2.2 Create histogram subplots to visualize all features

```python
fig, axes = plt.subplots(6, 2, figsize=(10, 30))
ndarray_red_features = df_red_wine.columns.values.reshape(6, 2)
for i in range(6):
    for j in range(2):
        str_feature = ndarray_red_features[i, j]
        df_red_wine[str_feature].plot(kind="hist", bins=30, ax=axes[i, j])
        axes[i, j].set(title=str_feature, ylabel="Count")
```
