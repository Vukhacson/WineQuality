```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_wine_data(path_to_dir, filename, csv_separator):
    csv_path = os.path.join(path_to_dir, filename)
    return pd.read_csv(csv_path, sep=csv_separator)
#os.path.dirname(path) return the parent directory of the given path
PARENT_DIR = os.path.dirname(os.getcwd())
WINE_DIR = os.path.join(PARENT_DIR, "Data set")
print(PARENT_DIR)
print(WINE_DIR)
```

```python
df_red_wine = load_wine_data(WINE_DIR, "winequality-red.csv", ";")
df_white_wine = load_wine_data(WINE_DIR, "winequality-white.csv", ";")
```

## 5. Correlation and Covariance

```python
df_red_cov_matrix = df_red_wine.cov()
df_white_cov_matrix = df_white_wine.cov()
```

```python
df_red_corr_matrix = df_red_wine.corr()
df_white_corr_matrix = df_white_wine.corr()
```

```python
df_red_cov_matrix.head()
```

```python
print("Red Wine Data: Covariance between quality and other features")
print(df_red_cov_matrix["quality"].sort_values(ascending=False))
print("\n\nWhite Wine Data: Covariance between quality and other features")
print(df_white_cov_matrix["quality"].sort_values(ascending=False))
```

```python
print("Red Wine Data: Correlation between quality and other features")
print(df_red_corr_matrix["quality"].sort_values(ascending=False))
print("\n\nWhite Wine Data: Correlation between quality and other features")
print(df_white_corr_matrix["quality"].sort_values(ascending=False))
```

```python
# Graph covariance & correlation using heat map
sns.heatmap(df_red_corr_matrix, cmap="Blues", linewidths=0.5, cbar=True)
```

```python
sns.heatmap(df_white_corr_matrix, cmap="Greys", linewidths=1)
```

# Deep Analysis

```python
np_red_features = df_red_wine.columns.values
np_red_features_reshaped = np_red_features.reshape(6, 2)
np_red_features_reshaped
```

## Visualize features to gain insights: Using boxplots

```python
fig, axes = plt.subplots(6, 2, figsize=(15, 25))
for i in range(6):
    for j in range(2):
        sns.boxplot(x="quality", y=np_red_features_reshaped[i, j], data=df_red_wine, ax=axes[i, j])
        axes[i, j].set_title(f"Boxplot of quality vs. {np_red_features_reshaped[i, j]}")
fig.tight_layout()
```

## Create histogram subplots to visualize all features

```python
fig, axes = plt.subplots(6, 2, figsize=(10, 30))
ndarray_red_features = df_red_wine.columns.values.reshape(6, 2)
for i in range(6):
    for j in range(2):
        str_feature = ndarray_red_features[i, j]
        df_red_wine[str_feature].plot(kind="hist", bins=30, ax=axes[i, j])
        axes[i, j].set(title=str_feature, ylabel="Count")
```
