```python
df_red_wine = pd.read_csv("G:\Learning resource\Github\WineQuality\Data set\winequality-red.csv", sep=';')
df_red_wine
```

```python
red_wine_feature = df_red_wine.columns.values
red_feature_reshaped = red_wine_feature.reshape(6,2)


red_feature = [0 for i in range(12)]

for i in range(12):
    red_feature[i] = df_red_wine.iloc[:,i].squeeze()
    
red_wine_feature
# red_feature_reshaped
```

```python
fig, axs = plt.subplots(6, 2, figsize = (15,45))

cmap = plt.get_cmap('viridis')
colors = cmap(np.linspace(4, 5, len(red_wine_feature)))

for i in range(6):
    for j in range(2):
        feature = red_feature_reshaped[i,j]
        axs[i,j].hist(df_red_wine[feature].values, bins=30)
        axs[i,j].set(title=feature, ylabel='Count')
```
