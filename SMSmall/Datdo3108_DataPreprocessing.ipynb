```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
def load_wine_data(path_to_dir, filename, csv_separator):
    csv_path = os.path.join(path_to_dir, filename)
    return pd.read_csv(csv_path, sep=csv_separator)
#os.path.dirname(path) return the parent directory of the given path
PARENT_DIR = os.path.dirname(os.getcwd())
WINE_DIR = os.path.join(PARENT_DIR, "Data set")
print(PARENT_DIR)
print(WINE_DIR)
```

```python
df_red_wine = load_wine_data(WINE_DIR, "winequality-red.csv", ";")
df_white_wine = load_wine_data(WINE_DIR, "winequality-white.csv", ";")
```

```python
# fixed acidity vs volatile acidity MinMaxScaler

df_red_wine_fixed_volatile = df_red_wine.iloc[:, [0, 1]]
df_red_wine_fixed_volatile

scaler = MinMaxScaler()
red_01_fit = scaler.fit_transform(df_red_wine_fixed_volatile)

print(red_01_fit)
type(red_01_fit)
```

```python
# original lineplot

plt.figure(figsize = (10,5))
plt.plot(np.linspace(0, 1, len(red_01_fit)), df_red_wine_fixed_volatile.iloc[:, 0], color='r')
plt.plot(np.linspace(0, 1, len(red_01_fit)), df_red_wine_fixed_volatile.iloc[:, 1], color='b')
plt.legend(['fixed', 'volatile'])
```

```python
# fitted lineplot

plt.figure(figsize = (10,5))
plt.plot(np.linspace(0, 1, len(red_01_fit)), red_01_fit[:, 0], color='r')
plt.plot(np.linspace(0, 1, len(red_01_fit)), red_01_fit[:, 1], color='b')
plt.legend(['fixed', 'volatile'])
```

```python
# original kdeplot

plt.figure(figsize = (10,5))
sns.kdeplot(df_red_wine_fixed_volatile.iloc[:, 0], color='r')
sns.kdeplot(df_red_wine_fixed_volatile.iloc[:, 1], color='b')
plt.legend(['fixed acidity', 'volatile acidity'])
```

```python
# fitted kdeplot

plt.figure(figsize = (10,5))
sns.kdeplot(red_01_fit[:, 0], color='r')
sns.kdeplot(red_01_fit[:, 1], color='b')
plt.legend(['fixed acidity', 'volatile acidity'])
```
