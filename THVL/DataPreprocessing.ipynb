**DATA SCALING**
- First, critical pre-processing step in DL
- Methods: normalizing and standardizing input and output variables

**WHY DATA SCALING**
- Unscaled input variables can result in a slow or unstable learning process
- Unscaled target variables on regression problems can result in exploding gradients causing the learning process to fail

# DEEP RESEARCH ON DATA SCALING

**RESEARCH: MaxAbsScaler**
- **Formula**:
                x_scaled = x / max(|x|)
- **Meaning of formula**: Scale each feature by its maximum absolute value. At first, the absolute maximum value of the feature is found and then the feature values are divided with it. 
- **How data transform**: Maximal absolute value of each feature in the training set will be 1. It does not shift/center the data, and thus does not destroy any sparsity.


**APPLY**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_redwine_csv = pd.read_csv("E:\LAB\ScaleMind\git\WineQuality\Data set\winequality-red.csv", sep = ';')
df_redwine_csv
```

Apply MaxAbsScaler

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
def plots(df, feature, trans):
    plt.figure(figsize = (13,5))
    plt.subplot(121)
    plt.hist(df[feature], bins=100)
    plt.title(feature +' before ' + str(trans).split('(')[0])

    plt.subplot(122)
    p1 = trans.fit_transform(df[[feature]])
    plt.hist(p1, bins =100)
    plt.title(feature + ' after '+ str(trans).split('(')[0])

for col in df_redwine_csv.columns:
    plots(df_redwine_csv, col , MaxAbsScaler())
```
