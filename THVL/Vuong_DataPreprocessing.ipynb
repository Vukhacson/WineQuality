Example of Data Scaling with red wine dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
```

```python
standard_scaler.fit(red_wine.drop('quality',axis=1))
```

```python
scaled_features = standard_scaler.transform(red_wine.drop('quality',axis=1))
```

```python
red_wine_after = pd.DataFrame(scaled_features,columns=red_wine.columns[:-1])
red_wine_after.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax_scaler = MinMaxScaler()
```

```python
minmax_scaler.fit(red_wine.drop('quality',axis=1))
```

```python
scaled_features1 = minmax_scaler.transform(red_wine.drop('quality',axis=1))
```

```python
red_wine_after1 = pd.DataFrame(scaled_features1,columns=red_wine.columns[:-1])
red_wine_after1
```
